```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize


def f1(x):
    return ((x % 6) ** 2 % 7) - np.sin(x)


# define the range for x where x is real numbers

x_values = range(1, 100)
y_values = [f1(x) for x in x_values]
# plottting to visualize the func behaviour
plt.plot(x_values, y_values)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) = -x^4 + 1000x^3 - 20x^2 + 4x - 6")
plt.grid(True)
plt.show()


results = [(x, f1(x)) for x in x_values]

# find the x that minimizes f(x)
x_min = max(results, key=lambda x: x[1])
print(x_min)
print(
    f" The value of x* that minimizes f(x) is x* = {x_min[0]}, with f1(x*) = {f1(x_min[1])}"
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize


# Define the function f(x)
def f(x):
    return -(x**4) + 1000 * x**3 - 20 * x**2 + 4 * x - 6


# Plot the function to visualize behavior over a wider range
x_values = np.linspace(-500, 1500, 1000)  # Updated range from -500 to 1500
y_values = f(x_values)

plt.plot(x_values, y_values)
plt.title("Plot of f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

# Multiple initial guesses to capture potential global minimum
initial_guesses = [0, 500, 1000, 1200]

# Store results
results = []

for guess in initial_guesses:
    result = minimize(f, x0=guess, method="BFGS")
    x_min = result.x[0]
    f_min = f(x_min)
    results.append((x_min, f_min))

# Find the global minimum across all results
global_min = min(results, key=lambda x: x[1])

print(f"The global minimum x* = {global_min[0]:.4f}, with f(x*) = {global_min[1]:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize


# Define the function f(x)
def f(x):
    return -(x**4) + 1000 * x**3 - 20 * x**2 + 4 * x - 6


# Define the negative of the function to find maxima
def neg_f(x):
    return -f(x)


# Plot the function to visualize behavior over a wider range
x_values = np.linspace(
    0, 1500, 1000
)  # Updated range from 0 to 1500 to capture the graph
y_values = f(x_values)

plt.plot(x_values, y_values)
plt.title("Plot of f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

# Limit the range of search based on visual inspection from the graph
bounds = [(500, 1000)]  # We limit the search around the maxima seen in the graph

# Use a new initial guess closer to the maxima and apply bounds
initial_guess = 700

# Store result using a constrained optimizer with bounds
result = minimize(neg_f, x0=initial_guess, bounds=bounds, method="L-BFGS-B")

# Extract x* and the function's maximum value
x_max = result.x[0]
f_max = f(x_max)  # Max value of original function (not negative)

print(f"The global maximum x* = {x_max:.4f}, with f(x*) = {f_max:.4f}")
```
